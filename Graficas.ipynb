####**Punto 1**

```python


def fibonacci(n):
    fib_series = [0, 1]
    for i in range(2, n):
        fib_series.append(fib_series[-1] + fib_series[-2])
    return fib_series[:n]

n = int(input("Ingrese el número de términos de Fibonacci: "))
fib_series = fibonacci(n)

plt.figure(figsize=(8, 5))
plt.plot(fib_series, marker='o', linestyle='-', color='b', label="Serie de Fibonacci")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.title("Serie de Fibonacci")
plt.legend()
plt.grid(True)
plt.show()


```

####**Punto 2**

```python

temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9,
                26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

media = statistics.mean(temperaturas)
mediana = statistics.median(temperaturas)
desviacion = statistics.stdev(temperaturas)

print(f"Temperatura media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Desviación estándar: {desviacion:.2f}")

plt.figure(figsize=(8, 5))
plt.plot(temperaturas, marker='o', linestyle='-', color='r', label="Temperatura diaria")
plt.axhline(media, color='b', linestyle='--', label="Temperatura media")
plt.xlabel("Día")
plt.ylabel("Temperatura (°C)")
plt.title("Temperatura diaria")
plt.legend()
plt.grid(True)
plt.show()


```

####**Punto 4**

```python
x = np.linspace(0, 2*np.pi, 1000)

y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

plt.plot(x, y_sin, label='Seno', color='blue')
plt.plot(x, y_cos, label='Coseno', color='red')
plt.plot(x, y_tan, label='Tangente', color='green')

plt.title('Funciones Trigonométricas: Seno, Coseno y Tangente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0, 2*np.pi])
plt.ylim([-2, 2])
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.show()
```

####**Punto 6**

```python
print("Ingrese las coordenadas de dos puntos en el plano.")
x1 = float(input("x1: "))
y1 = float(input("y1: "))
x2 = float(input("x2: "))
y2 = float(input("y2: "))

if x2 == x1:
    print("La recta es vertical y la pendiente tiende a infinito.")

    x_values = np.array([x1, x1])
    y_values = np.linspace(min(y1, y2)-5, max(y1, y2)+5, 100)


    plt.plot(x_values, y_values, label=f"Recta vertical x = {x1}", color='blue')

    plt.scatter([x1, x2], [y1, y2], color='red', label="Puntos dados")

    plt.title("Gráfico de la recta vertical a partir de dos puntos")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axhline(0, color='black', linewidth=0.8)
    plt.axvline(0, color='black', linewidth=0.8)
    plt.grid(True)
    plt.legend()
    plt.show()

else:

    m = (y2 - y1) / (x2 - x1)

    b = y1 - m * x1

    print(f"La pendiente de la recta es: {m}")
    print(f"La ecuación de la recta es: y = {m}x + {b}")


    x_values = np.linspace(min(x1, x2)-1, max(x1, x2)+1, 100)
    y_values = m * x_values + b


    plt.plot(x_values, y_values, label=f"Recta: y = {m:.2f}x + {b:.2f}", color='blue')

    plt.scatter([x1, x2], [y1, y2], color='red', label="Puntos dados")

    plt.title("Gráfico de la recta a partir de dos puntos")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axhline(0, color='black', linewidth=0.8)
    plt.axvline(0, color='black', linewidth=0.8)
    plt.grid(True)
    plt.legend()
    plt.show()
```

####**Punto 7**

```python
def movimiento_parabolico(Vo, angulo, g=9.8):
    angulo = np.radians(angulo)
    t = np.linspace(0, 2 * Vo * np.sin(angulo) / g, 10000)
    return Vo * np.cos(angulo) * t, Vo * np.sin(angulo) * t - 0.5 * g * t**2

plt.figure(figsize=(10, 7))

Vo= 30

for angulo in [30, 40, 50]:
    plt.plot(*movimiento_parabolico(Vo, angulo), label=f"Angulo = {angulo}°, Vo = {Vo} m/s")

plt.xlabel("Distancia horizontal (m)")
plt.ylabel("Distancia vertical (m)")
plt.title("Movimiento parabolico (proyectil)")
plt.grid(True)
plt.legend()
plt.show()
```

####**Punto 8**

```python
def fun_x_n(n):
  x = np.linspace(0, 10, 1000)
  fun = x**n

  plt.plot(x, fun, color="g", linestyle='-')
  plt.title("Gráfica de la función $f(x) = x^n$")
  plt.xlabel("Eje X")
  plt.ylabel("Eje Y")
  plt.grid(True)
  plt.show()

slider = widgets.IntSlider(
    value=1,
    min=1,
    max=20,
    step=1,
    description="n ="
)

widgets.interactive(fun_x_n, n=slider)
```

####**Punto 9 (extra)**

```python
num_points = 10000

corners = np.array([
    [0, 0],
    [1, 0],
    [0.5, np.sqrt(3) / 2]
])


p = np.random.rand(2)
points = []


i = 0
while i < num_points:

    corner = corners[np.random.randint(0, 3)]

    p = (p + corner) / 2

    points.append(p)

    i += 1

points = np.array(points)

plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0], points[:, 1], s=0.2, color='black')
plt.title("Fractal de Sierpinski con Chaos Game")
plt.axis('equal')
plt.axis('off')
plt.show()
```
