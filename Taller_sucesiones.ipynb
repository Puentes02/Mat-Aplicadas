# ¿Tiene límite la secuencia $( a_n = 2^n )$?

La secuencia dada es:

$[a_1 = 2, \quad a_{n+1} = 2a_n]$

Podemos encontrar la fórmula general:

$[a_n = 2^a]$

### Análisis del límite

Para determinar si la secuencia tiene un límite, evaluamos:

$[\lim_{n \to \infty} 2^n]$

Sabemos que la función exponencial \( 2^n \) crece rápidamente conforme \( n \) aumenta. Es decir:

$[2^n \to \infty \quad \text{cuando } n \to \infty]$

Dado que la secuencia crece indefinidamente, **no es acotada** y **no converge a un número finito**.

### **Conclusión**
Como $( a_n )$ tiende a infinito, la secuencia **no tiene límite**.

```python
n_values = np.arange(1, 51)
a_values = [10]

for n in range(1, 50):
    a_values.append(a_values[-1] / 3)

plt.figure(figsize=(10, 5))
plt.plot(n_values, a_values, marker="o", linestyle="-", color="blue", label=r"$a_n = \frac{1}{3} a_{n-1}$")

plt.xlabel("n")
plt.ylabel("a_n")
plt.title("Gráfica de la secuencia a_n")
plt.legend()
plt.grid()


plt.show()
```

La secuencia dada es:

$[a_1 = 10, \quad a_{n+1} = \frac{1}{3} a_n]$

Podemos encontrar la fórmula general aplicando recursión:

$[a_2 = \frac{1}{3} a_1 = \frac{10}{3}]$

$[a_3 = \frac{1}{3} a_2 = \frac{10}{9}]$

De forma general:

$[a_n = 10 \left(\frac{1}{3}\right)^{n-1}]$

### **Análisis del límite**

Queremos determinar:

$[\lim_{n \to \infty} a_n = \lim_{n \to \infty} 10 \left(\frac{1}{3}\right)^{n-1}]$

Dado que $( \frac{1}{3} )$ es una fracción menor que 1, cuando $( n )$ tiende a infinito:

$[\left(\frac{1}{3}\right)^{n-1} \to 0]$

Por lo tanto:

$[a_n \to 10 \cdot 0 = 0]$

### **Conclusión**
La secuencia es **decreciente y acotada** (nunca es negativa) y **converge a 0**.

```python
# Definir los valores de n
n_values = np.arange(1, 101)

# Definir las secuencias
a_values = n_values  # a_n = n
b_values = -n_values  # b_n = -n
sum_values = a_values + b_values  # Suma de las secuencias

# Graficar las secuencias
plt.figure(figsize=(10, 5))
plt.plot(n_values, a_values, label=r"$a_n = n$", color="red")
plt.plot(n_values, b_values, label=r"$b_n = -n$", color="blue")
plt.plot(n_values, sum_values, label=r"$a_n + b_n = 0$", color="green", linestyle="dashed")

# Configuración del gráfico
plt.xlabel("n")
plt.ylabel("Valor de la secuencia")
plt.title("Ejemplo de secuencias divergentes cuya suma converge")
plt.axhline(y=0, color='black', linestyle="dotted")  # Línea horizontal en y=0
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()
```

- $a_n = n$ (diverge a $(+\infty)$).
- $b_n = -n$ (diverge a $(-\infty)$).

Claramente, ambas secuencias divergen, pero su suma es:

$[a_n + b_n = n + (-n) = 0]$

Esta secuencia es constante y converge a 0.

 **Movimiento de una Pelota en Rebotes**


 **Paso 1: Definir la secuencia de alturas**
Cada rebote alcanza una altura que sigue la progresión geométrica:

$$
h_n = 10 \times (0.75)^n
$$

distancia total recorrida es la suma de la altura de la caída inicial más dos veces la suma infinita de los rebotes:

$$
\text{Distancia total} = 10 + 2 \sum_{n=1}^{\infty} 10 (0.75)^n
$$

 **Paso 2: Calcular la suma infinita**
razón \( r = 0.75 \) y primer término \( a = 10(0.75) \):

$$
S = \frac{a}{1 - r} = \frac{10(0.75)}{1 - 0.75} = \frac{7.5}{0.25} = 30
$$

 **Paso 3: Obtener la distancia total**


$$
\text{Distancia total} = 10 + 2(30) = 10 + 60 = 70 \text{ metros}
$$

Por lo tanto, la pelota recorre **70 metros** antes de detenerse.

---

Si la pelota rebotara solo al **25%** de su altura anterior, seguimos el mismo proceso pero con \( r = 0.25 \):

$$
S = \frac{10(0.25)}{1 - 0.25} = \frac{2.5}{0.75} \approx 3.33
$$

Entonces:

$$
\text{Distancia total} = 10 + 2(3.33) = 10 + 6.67 = 16.67 \text{ metros}
$$

En este caso, la pelota recorrería aproximadamente **16.67 metros** antes de detenerse.

$$
H_n = \sum_{k=1}^{n} \frac{1}{k}
$$

Para encontrar el valor de \( n \) que satisface \( H_n > 5 \), sumamos los términos hasta que la condición se cumpla:

$$
H_{83} \approx 5.0002
$$

Por lo tanto, se necesitan al menos **83** términos para que la suma parcial supere 5.

Para encontrar el valor de \( n \) que satisface \( H_n > 10 \):

$$
H_{12367} \approx 10.0001
$$

Por lo tanto, se necesitan al menos **12,367** términos para que la suma parcial supere 10.
